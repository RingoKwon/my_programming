```python

import awswrangler as wr
import pandas as pd
# import pymysql
# import datetime

    

# Lambda 핸들러 함수. 이벤트와 컨텍스트 매개변수를 받습니다.


# 첫 번째 SQL 쿼리 실행
sql_query_all = """
select
id
, status
,reg_ts
, mvno
, order_phone_number
, customer_phone_number
, customer_name
, completed_at 
, plan_name
from moyo_order.mobile_plan_order a
where 1=1
    and a.mvno = '미니게이트'
    and a.reg_ts > date( '2023-05-01')
"""

df = wr.athena.read_sql_query(sql=sql_query_all, database="moyo_logs")
```

```python
df_mini_out = pd.read_csv('mvno 정산자료 검증 - 미니게이트 6월.csv')
```

```python
df_mini_out.head()
```

```python
df_mini_out['가입자 번호'] = '0'+df_mini_out['가입자 번호'].astype(str)
```

```python
df_mini_out.head()
```

```python
result = pd.merge(df, df_mini_out, how='outer', left_on='order_phone_number', right_on='가입자 번호')
```

```python
# export to csv
result.to_csv('mvno 정산자료 검증 - 미니게이트 6월 v1.csv', index=False)
```
